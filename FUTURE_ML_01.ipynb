```python

import pandas as pd
from prophet import Prophet
import logging


logging.getLogger('cmdstanpy').setLevel(logging.WARNING) 


df = pd.read_csv('Sample - Superstore.csv', encoding='ISO-8859-1')


df['Order Date'] = pd.to_datetime(df['Order Date'])


print("Data loaded! Here are the first 3 rows:")
df.head(3)
```

```python

all_results = []


categories = df['Category'].unique()

print("Starting the AI training... (This usually takes 1-2 minutes)")

for cat in categories:
    print(f"Processing: {cat}...")
    
    
    cat_df = df[df['Category'] == cat]
    
   
    prophet_data = cat_df.set_index('Order Date').resample('MS')['Sales'].sum().reset_index()
    prophet_data.columns = ['ds', 'y']
    
   
    model = Prophet(interval_width=0.95, weekly_seasonality=False, daily_seasonality=False)
    model.add_seasonality(name='yearly', period=365.25, fourier_order=5)
    model.fit(prophet_data)
    
    
    future_dates = model.make_future_dataframe(periods=24, freq='MS')
    forecast = model.predict(future_dates)
    
    
    
    
    actual_data = prophet_data.copy()
    actual_data['Type'] = 'Actual'
    actual_data['Category'] = cat
    actual_data = actual_data.rename(columns={'ds': 'Order Date', 'y': 'Sales'})
    
    
    last_real_date = prophet_data['ds'].max()
    future_only = forecast[forecast['ds'] > last_real_date].copy()
    
    forecast_data = pd.DataFrame({
        'Order Date': future_only['ds'],
        'Sales': future_only['yhat'], 
        'Category': cat,
        'Type': 'Forecast',
        'Lower Bound': future_only['yhat_lower'], 
        'Upper Bound': future_only['yhat_upper']  
    })
    
    
    all_results.append(actual_data)
    all_results.append(forecast_data)

print("Finished! All categories processed.")
```

```python

final_df = pd.concat(all_results, ignore_index=True)

final_df.to_csv('Prophet_Forecast_Output.csv', index=False)

print("SUCCESS! File 'Prophet_Forecast_Output.csv' has been created.")
print("Go to your folder and check if it's there.")
```
